# Building Neural Net

```python
import numpy as np
import matplotlib.pylab as plt
```

## Activation Functions

### Step Function

```python
def step_function(x):
    return np.array(x>0)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.xlim(-6.0, 6.0)
plt.show()
```

### Sigmoid

```python
def sigmoid(x):
    return 1/ (1 + np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)

plt.plot(x,y)
plt.ylim(-0.1, 1.1)
plt.xlim(-6.0, 6.0)
plt.show()
```

### ReLU

```python
def relu(x):
    return np.maximum(0,x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

plt.plot(x,y)
plt.xlim(-6.0, 6.0)
plt.show()
```
